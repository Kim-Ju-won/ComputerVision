## 영상의 회전

- warpAffine

```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("china.jpg")
r, c = img.shape[:2]
M = cv.getRotationMatrix2D((c/2,r/2),60,1)
new_img = cv.warpAffine(img, M, (c,r))

plt.imshow(new_img)
plt.plot()
```

## Linear interpolation

- Color interpolation

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def color_rect():

    # Color rectangle 생성
 
    g = np.zeros((256,256,3), np.uint8)
 
    for i in range(256):
        g[i,:,:] = (0, i, 255-i)

    return g

g = color_rect()

#cv.imshow('Org', g)
col = cv.cvtColor(g, cv.COLOR_BGR2RGB)
plt.imshow(col)
plt.show()
```
