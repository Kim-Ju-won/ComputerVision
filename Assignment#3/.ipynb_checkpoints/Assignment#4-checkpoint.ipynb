## 1. 실습 2 : ORB를 이용한 정합 함수 - 소스 코드 및 실행 결과

- ORB로 추출한 특징점들에 대해 기술자(descriptor)를 계산: orb.detectAndCompute
- 계산된 기술자를 brute force matching: cv.BFMatcher
- 유사도를 기준으로 결과 순서를 조절
- 결과를 함수로 영상에 표시 : cv.drawMatches
  - 이 경우 유사도를 기준으로 정합의 숫자를 조절할 수 있음(Matches조절)

````python
import cv2 as cv
import matplotlib.pyplot as plt

img1 = cv.imread('b1.png',cv.IMREAD_GRAYSCALE)  # queryImage
img2 = cv.imread('b2.png',cv.IMREAD_GRAYSCALE) 	# trainImage

# Initiate ORB detector
orb = cv.ORB_create()
# find the keypoints and descriptors with ORB
kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)

# create BFMatcher object
bf = cv.BFMatcher(cv.NORM_HAMMING, crossCheck=True)
# Match descriptors.
matches = bf.match(des1,des2)
# Sort them in the order of their distance.
matches = sorted(matches, key = lambda x:x.distance)
# Draw first 30 matches.
img3 = cv.drawMatches(img1,kp1,img2,kp2,matches[:100],None,
                      flags=cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
# 크기를 조금더 크게 보여주기 위해 코드를 추가하였음
plt.figure(figsize=(17,17)) 
plt.imshow(img3,'gray')
plt.show()
````

## 2. 실습 3: 물체 인식 - 소스코드 및 실행 화면
- 모델 영상과 장면 영상을 비교하여 장면 영상에서 모델을 찾음
- 앞 프로그램 중 입력 영상을 수정하여 실행 시킴
    - b1.png와 b2.png를 각각 p1.jpg와 p2.jpg로 수정

````python

import cv2 as cv
import matplotlib.pyplot as plt

img1 = cv.imread('p1.jpg',cv.IMREAD_GRAYSCALE)  # queryImage
img2 = cv.imread('p2.jpg',cv.IMREAD_GRAYSCALE) 	# trainImage

# Initiate ORB detector
orb = cv.ORB_create()
# find the keypoints and descriptors with ORB
kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)

# create BFMatcher object
bf = cv.BFMatcher(cv.NORM_HAMMING, crossCheck=True)
# Match descriptors.
matches = bf.match(des1,des2)
# Sort them in the order of their distance.
matches = sorted(matches, key = lambda x:x.distance)
# Draw first 30 matches.
img3 = cv.drawMatches(img1,kp1,img2,kp2,matches[:30],None,
                      flags=cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
# 크기를 조금더 크게 보여주기 위해 코드를 추가하였음
plt.figure(figsize=(17,17)) 
plt.imshow(img3,'gray')
plt.show()
````
