# openCV 예제 1

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('lena512.tif')

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

x_edges = cv.Sobel(gray, -1, 1, 0, ksize=5)
y_edges = cv.Sobel(gray, -1, 0, 1, ksize=5)

plt.subplot(1,3,1)
plt.imshow(img)
plt.subplot(1,3,2)
plt.imshow(x_edges)
plt.subplot(1,3,3)
plt.imshow(y_edges)
```

# OpenCV 예제 2

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt


img = cv.imread('lena512.tif') 
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

edges = cv.Canny(gray, 50, 100, 3)

plt.imshow(edges)
plt.show()
```
