### 1. 실습 1~2 결과를 제시
#### 1.1 실습 1결과

````python
# 사진 확인
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

i0 = cv.imread('im2.png')
i1 = cv.imread('im6.png')
plt.subplot(1,2,1)
plt.imshow(i0)
plt.subplot(1,2,2)
plt.imshow(i1)
````

- 함수 정의 : computeMSD

````python
# 함수 정의
def computeMSD(imgL, imgR, x, y, dispRange):
    N = 7
    # ref는 im0에서의 (2N+1)x(2N+1) 기준 영역
    ref = imgL[y-N:y+N+1, x-N:x+N+1]
 
    # MSD는 각 disparity 위치에서의 MSD 값을 저장
    MSD = np.zeros(dispRange)
    for k in range(dispRange):
        # targe은 im1에서의 (2N+1)x(2N+1) 정합 영역
        target = imgR[y-N:y+N+1, x-k-N:x-k+N+1]
        # Diff는 기준 영역과 정합영역의 차이
        Diff = ref - target
        MSD[k] = np.mean(Diff*Diff)
    # MSD 최소값 위치가 disparity
    disparity =  np.argmin(MSD)
    return MSD, disparity
````

- 실습 1 함수 실행 결과 : 

````python
MSD, disp = computeMSD(i0, i1, 210, 279, 64)
plt.plot(MSD)
print('Disparity: ', disp)
````

### 2. computeMSD 함수를 이용하여 stereo.compute을 대체하는 함수를 작성하고자 함. depthMap을 계산하는 함수(다음 장)를 완성하고 이 함수를 이용하여 실습 2를 다시 수행한 결과를 제시함 (주의: 이렇게 완성된 함수는 수행 시간이 1분 이상이 될 수 있으니 결과를 얻기까지 충분히 기다릴 것) 

- 함수 정의 
    - computeMSD를 이용한 disparityMSD함수 정의

````python
from tqdm.notebook import tqdm as t # 시간확인용 변수

def disparityMSD(imgL, imgR, dispRange):
    row, col = imgL.shape
    disparityMap = np.zeros((row, col))
    
    for y in t(range(11, row-10)):
        for x in range(74,col-10):
            # 함수로 받아온 값에서 disp값만 활용
            MSD, disp = computeMSD(imgL,imgR, x, y, dispRange)
            disparityMap[y,x] = disp
    return disparityMap
````

- 함수를 적용해서 출력한 코드

````python
# 실습2를 
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

imgL=cv.imread('im2.png')
imgR=cv.imread('im6.png')

# 영상을 gray로 변환
gray1 = cv.cvtColor(imgL, cv.COLOR_BGR2GRAY)
gray2 = cv.cvtColor(imgR, cv.COLOR_BGR2GRAY)

disparityMap = disparityMSD(gray1, gray2, 64)
plt.imshow(disparityMap,'gray')
plt.show()
````
